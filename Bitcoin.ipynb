```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.read_csv('bitcoin.csv')
```

```python
bitcoin=pd.read_csv('bitcoin.csv')
type(bitcoin)
```

```python
plt.plot(bitcoin.Open, bitcoin.Volume, '-')

plt.title("Bitcoin open-volume USD")
plt.xlabel('Open')
plt.ylabel('Volume')

plt.show()

plt.plot(bitcoin.Close, bitcoin.Takerbuybaseassetvolume, '-')
plt.title("Bitcoin close-volume USD")
plt.xlabel('Close')
plt.ylabel('Taker buy base asset volume')

plt.show()
```

```python
bitcoin[bitcoin.Takerbuyquoteassetvolume >= 5000000]

```

```python
high = bitcoin.High
low = bitcoin.Low
plt.plot(bitcoin.Open, bitcoin.High, '*', label = 'high')
plt.plot(bitcoin.Open, bitcoin.Low, 'o', label ='low')

plt.title("Bitcoin open-volume USD")
plt.xlabel('Open')
plt.ylabel('Volume')

plt.legend()

plt.show()

high = bitcoin.High
low = bitcoin.Low
plt.plot(bitcoin.Close, bitcoin.High, '*', label = 'high')
plt.plot(bitcoin.Close, bitcoin.Low, 'x', label ='low')

plt.title("Bitcoin close-volume USD")
plt.xlabel('Close')
plt.ylabel('Takerbuybaseassetvolume')

plt.legend()

plt.show()
```
